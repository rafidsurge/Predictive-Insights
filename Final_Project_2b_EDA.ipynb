```python
df2 = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/ChampionStats2.csv')
```

```python
champion_stats_cleaned = df2.copy()
```

```python
columns_to_convert = ['KP', 'CS%P15', 'DMG%', 'GOLD%', 'W%']
```

```python
for col in columns_to_convert:
    champion_stats_cleaned[col] = pd.to_numeric(champion_stats_cleaned[col].str.rstrip('%'), errors='coerce') / 100.0
```

```python
features = ['KDA', 'KP', 'GD10', 'XPD10', 'CSD10', 'CSPM', 'CS%P15', 'DPM', 'DMG%', 'GOLD%', 'WPM', 'WCPM']
for col in features:
    champion_stats_cleaned[col] = pd.to_numeric(champion_stats_cleaned[col], errors='coerce')
```

```python
champion_stats_cleaned['GP'] = pd.to_numeric(champion_stats_cleaned['GP'], errors='coerce')
```

```python
champion_stats_cleaned.dropna(inplace=True)
```

```python
columns = ['W%','K','D','A','KDA', 'KP', 'GD10', 'XPD10', 'CSD10', 'CSPM', 'CS%P15', 'DPM', 'DMG%', 'GOLD%', 'WPM', 'WCPM']
```

```python
plt.figure(figsize=(20, 15))
for i, col in enumerate(columns):
    if col in champion_stats_cleaned.columns:  # Ensure the column exists before plotting
        plt.subplot(4, 4, i + 1)
        sns.boxplot(y=champion_stats_cleaned[col])
        plt.title(f'Boxplot of {col}')
        plt.tight_layout()
    else:
        print(f"Column '{col}' does not exist in the DataFrame.")

plt.show()
```
