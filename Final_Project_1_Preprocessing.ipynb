##Data Preprocessing

Checking the five files for column similarities

```python
dataframes = [df1, df2, df3, df4, df5]
```

```python
columns_list = [df.columns for df in dataframes]
```

```python
same_columns = all(columns_list[0].equals(columns) for columns in columns_list)
```

```python
if same_columns:
    print("All DataFrames have the same columns.")
else:
    print("The DataFrames do not have the same columns.")
    for i, columns in enumerate(columns_list):
        print(f"Columns in DataFrame {i+1}: {list(columns)}")
```

As the five files have same columns, we will now combine the files

```python
combined_df = pd.concat(dataframes, ignore_index=True)
```

```python
combined_df.head()
```

```python
combined_df.shape[0]
```

Keeping only the major league match data (CBLOL, LCK, LEC, LCS, LJL, LLA, LPL, PCS,VCS)

```python
leagues_to_keep = ['CBLOL', 'LCK', 'LEC', 'LJL', 'LLA', 'LPL', 'LCS', 'PCS', 'VCS']
```

```python
combined_df = combined_df[combined_df['league'].isin(leagues_to_keep)]
```

```python
combined_df.shape[0]
```

This combined_df contains both match and player data. we will segregate the teams data and player data onto different dataset.

```python
player_df = combined_df[combined_df['position'] != 'team']
```

```python
player_df.describe()
```

```python
player_df['position'].unique()
```

```python
player_df.shape[0]
```

```python
player_df.isna().sum()
```

Keeping only the relevant columns that we may or may not use depending on further test.

```python
columns_to_keep = [
    'position', 'champion', 'result', 'kills', 'deaths', 'assists', 'teamkills', 'teamdeaths', 'firstblood', 'firstbloodkill', 'firstbloodassist', 'firstbloodvictim',
    'team kpm', 'ckpm', 'damagetochampions', 'dpm', 'damageshare', 'damagetakenperminute', 'damagemitigatedperminute', 'wardsplaced', 'wpm', 'wardskilled', 'wcpm',
    'controlwardsbought', 'visionscore', 'vspm', 'totalgold', 'earnedgold', 'earned gpm', 'earnedgoldshare', 'goldspent', 'total cs', 'minionkills', 'monsterkills',
    'cspm', 'goldat10', 'xpat10', 'csat10', 'opp_goldat10', 'opp_xpat10', 'opp_csat10', 'golddiffat10', 'xpdiffat10', 'csdiffat10', 'killsat10', 'assistsat10', 'deathsat10',
    'opp_killsat10', 'opp_assistsat10', 'opp_deathsat10', 'goldat15', 'xpat15', 'csat15', 'opp_goldat15', 'opp_xpat15', 'opp_csat15', 'golddiffat15', 'xpdiffat15', 'csdiffat15',
    'killsat15', 'assistsat15', 'deathsat15', 'opp_killsat15', 'opp_assistsat15', 'opp_deathsat15'
]
```

```python
clean_df = player_df[columns_to_keep]
```

```python
clean_df.info()
```

```python
clean_df.shape[0]
```

```python
clean_df.isna().sum()
```

As per the Dataset provider, and the LOL esports official website, there are instances where a game was cancelled mid game due to bug. But because the game was started, the records were registered nonetheless. Due to the nature of the dataset, Filling the null values with regular preprocessing method might actually give us deminishing returns. instead, just dropping the rows containing null values. That still leaves us with a very large dataset to work with.

```python
final_df = clean_df.dropna()
```

```python
final_df.isna().sum()
```

```python
final_df.to_csv('Final players df.csv', index=False)
```
